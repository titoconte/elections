# Definitions

1. split_cd - Used to split the list with strings separeted by ";". It's made to separete string with section numbers in arquivo.csv

2. cros_election_data - Used to insert school names in elections data frame (code corresponds) and return a grouped dataframe with vote number by scholl, candidate, zone, city and position.

 1. I get the section number and zone number and search for corresponded number;
 2. Change school number by school name;
 3. Grouped the information with sum based in location number

```python
# function to split sections
split_cd = lambda x:x.split(';')

# definition to process election data
def cross_election_data(fname,cols):

	# load data
	df = pd.read_csv(
        fname,
        # brazilian csv
        sep=';',
        # no header
        header=None,
        # solve ecoding problems
        encoding='ISO-8859-1'
    )	

	# inser colums names
	df.columns=cols

	# inserts schools names
	df['ESCOLAS']=np.zeros(df.shape[0])
	# gets the section information - same requisites as fname load
	dfplace=pd.read_csv(
        'arquivo.csv',
        header=None,
        encoding='ISO-8859-1'
    )    

	# gets the section number
	sections = list(dfplace.iloc[:,-1])

	print('splited')
	# converts selections to list (applies split_cd function) because the sections numbers
    # are a unique strings separeted by ";"
	sections = list(map(split_cd,sections))

	print('inserting schools')
	# gets max zone number
	maxzone = dfplace.iloc[:,0].max()
	zzone=0
	# replace school names
	for lsection,zone,school in zip(sections,dfplace.iloc[:,0],dfplace.iloc[:,2]): 
        
        # check zone and print progress messange
		if zzone<zone:
			print('processing zone {} of {}'.format(zone,maxzone))
			zzone=zone
            
		# gets section number sample
		if '-' in lsection[0]:
			sec = lsection[0].split('-')[1]
		else:
			sec = lsection[0]
            
		# gets local number
		local = df.loc[(df['ZONA']==int(zone)) & (df['SECAO']==int(sec)),'LOCAL']
		if len(local)>0:
			# inserts school number based in local number
			df.loc[df['LOCAL']==local.unique()[0],'ESCOLAS']=school
		else:
			# shows some error
			print('school {} not found'.format(school))
            
	print('schools inserted')

	print('format converted')
	# group by selections information
	dfsum = df.groupby(['NOME_DO_MUNICIPIO','ZONA','CARGO','ESCOLAS','CANDIDATO'],axis=0).sum()
    # gets interested values (just votes)
	dfsum=dfsum.loc[:,'VOTOS']
	# export csv with result
	dfsum.to_csv(fname.replace('.txt','.csv'))
	print('grouped')
```

# Intermediate process 
This part loads infos and create the intermediate files. This approach is because cross_election_data spend mutch time to run and the result change based in use input.
The dfsum is a first result to plot

```python
# gets zones numbers same as last code block
zonas = np.unique(pd.read_csv(
    'arquivo.csv',
    header=None,
    encoding='ISO-8859-1'
    # gets the first column with zone number
).loc[:,0])
# if the intermediate file doesn't exist
if fname.replace('.txt','.csv') not in os.listdir():
    # applies function to get data
    df = cross_election_data(fname,cols)

# loads the result from cross_election_data function
df = pd.read_csv(fname.replace('.txt','.csv'),header=None)
# creates df header
COLS = ['NOME_DO_MUNICIPIO','ZONA','CARGO','ESCOLAS','CANDIDATO','VOTOS']
df.columns = COLS
# creates a first df result
dfsum = df.loc[(df.ZONA==zonas[0]) & (df.CARGO=='Prefeito'),:]
```

## Bokeh heatmap plot and update definition

```python
# to heat map forget about it
# loads bokeh functions
from bokeh.models import ColumnDataSource,LinearColorMapper, FactorRange
from bokeh.io import output_notebook, show,push_notebook
from bokeh.plotting import figure
from bokeh.palettes import YlOrRd9

# creates a output noteboo
output_notebook()
# gets source info from dataframe
source = ColumnDataSource(dfsum)
# creates heatmap
hm = figure(
    # infos of to axis
    x_range=list(dfsum.CANDIDATO.unique()),
    y_range=list(dfsum.ESCOLAS.unique())
)
# creates color range
mapper = LinearColorMapper(low=dfsum.VOTOS.min(),high = dfsum.VOTOS.max(),palette=YlOrRd9)
# applies colorpalette to field
fill_color={'field': 'VOTOS','transform': mapper}
# create heatmap
hm.rect(source=source,x='CANDIDATO',y='ESCOLAS',width=1, height=1,fill_color=fill_color,line_color=None)
# rotate names in heatmap x-axis
hm.xaxis.major_label_orientation = np.pi/4

```

```python
# plot
show(hm,notebook_handle=True)
```
